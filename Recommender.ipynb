```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter = counter + 1
        else:
            break
    return L
   
```

```python
def convert3(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
import re
```

```python
nltk.download("stopwords")
```

```python
ps = PorterStemmer()
```

```python
def cleaning(tags):
    cleaned = re.sub("[^a-zA-Z0-9]"," ",tags)
    cleaned = cleaned.lower()
    cleaned = cleaned.split()
    cleaned = [ps.stem(word) for word in cleaned]
    cleaned = " ".join(cleaned)
    return cleaned
```

```python
movies["Tags"] = movies["Tags"].apply(cleaning)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tdf = TfidfVectorizer(stop_words="english")
```

```python
vector = tdf.fit_transform(movies["Tags"]).toarray()
```

```python
vector[1]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = int(movies[movies["title"] == movie].index[0])
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse= True, key= lambda x:x[1])[1:6]

    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
